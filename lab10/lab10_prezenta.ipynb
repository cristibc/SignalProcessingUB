```python
import math
import random

def generate_normal_distribution(mean, std_dev, size=1):
    samples = []

    for _ in range(size):
        u1, u2 = random.random(), random.random()
        z0 = math.sqrt(-2.0 * math.log(u1)) * math.cos(2.0 * math.pi * u2)
        sample = mean + std_dev * z0
        samples.append(sample)

    return samples

mean_value = 0.0
std_deviation = 1.0
size_of_samples = 1000

result = generate_normal_distribution(mean_value, std_deviation, size_of_samples)
print(result)
N = 1000
t = np.arange(N)

plt.hist(result)
plt.show()

mean = (0, 0)
cov = [[6, -3], [-3, 5]]
x, y = random.multivariate_normal(mean, cov, 1000).T
plt.scatter(x, y)
plt.show()

def multivariate_normal(x, d, mean, covariance):
    """pdf of the multivariate normal distribution."""
    x_m = x - mean
    return (1. / (np.sqrt((2 * np.pi)**d * np.linalg.det(covariance))) * 
            np.exp(-(np.linalg.solve(covariance, x_m).T.dot(x_m)) / 2))

```
